# DBSCAN Clustering - Density Based
## for set of random points

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Generating Random points

```python
from sklearn.datasets import samples_generator

n_samples = 1500
n_features = 2
centers = [[0, 0], [1, 4], [5, 0], [4, 3]]   # Centroids for k = 4
data, y = samples_generator.make_blobs(n_samples, n_features, centers, cluster_std=0.5)

plt.scatter(data[:,0], data[:,1], c='b', marker='o')
plt.show()
```

### Training DBSCAN

```python
from sklearn import cluster

R = 0.3
M = 7

clust = cluster.DBSCAN(eps=R, min_samples=M)
clust.fit(data)

labels = clust.labels_

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

### Ploting Clusters

```python
plt.scatter(data[:, 0], data[:, 1], c=labels)
plt.show()
```
