# <center>Demonstration of Various Functions</center>

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Linear Function

#### $$ y = mx + c $$

```python
x = np.arange(-5, 5, 0.1)
y = 2 * x + 3
y_noise = 2 * np.random.randn(len(x))
y_data = y + y_noise

plt.scatter(x, y_data, c='r')
plt.plot(x, y, c='b')
plt.show()
```

## Non Linear Functions
### Quadratic Function
#### $$ y = ax^2 + bx + c $$

```python
y = 3 * (x**2) + 3 * x + 4
y_noise = 5 * np.random.randn(len(x))
y_data = y + y_noise

plt.scatter(x, y_data, c='r')
plt.plot(x, y, c='b')
plt.show()
```

### Cubic Equation
#### $$ y = ax^3 + bx^2 + cx + d $$

```python
y = 1 * (x**3) + 2 * (x**2) + 3 * x + 3
y_noise = 10 * np.random.randn(len(x))
y_data = y + y_noise

plt.scatter(x, y_data, c='r')
plt.plot(x, y, c='b')
plt.show()
```

### Exponential Function
#### $$ y = a + bc^x $$

```python
y = 2 + 3 * (2**x)
y_noise = 5 * np.random.randn(len(x))
y_data = y + y_noise

plt.scatter(x, y_data, c='r')
plt.plot(x, y, c='b')
plt.show()
```

### Logarithmic Function
#### $$ y = Log(x) $$

```python
y = np.log(x)
y_noise = 0.2 * np.random.randn(len(x))
y_data = y + y_noise

plt.scatter(x, y_data, c='r')
plt.plot(x, y, c='b')
plt.show()
```

### Sigmoidal/Logistic
#### $$ y = a + \frac{b}{1 + c^{x - d}} $$

```python
y = 1 - 1 / (1 + np.power(4, x - 1))
y_noise = 0.1 * np.random.randn(len(x))
y_data = y + y_noise

plt.scatter(x, y_data, c='r')
plt.plot(x, y, c='b')
plt.show()
```
